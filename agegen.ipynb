```python
import numpy as np
import pandas as pd
from PIL import Image
import cv2
import os
import matplotlib.pyplot as plt
from tqdm import tqdm


dir = r'C:\Users\Kartik\Desktop\agegender dataset\kaggle\\'

df_age = pd.read_csv(dir + 'train_age.csv')
df_gen = pd.read_csv(dir + 'train_gender.csv')
print(df_age.head())
print(df_gen.head())
im = []
# Correct path concatenation
train_images_dir = os.path.join(dir, 'kaggle', 'train')  # Use os.path.join to handle path correctly

# Ensure the images are present in the specified directory
if not os.path.exists(train_images_dir):
    print(f"Error: The directory {train_images_dir} does not exist.")
else:
    for i in tqdm(sorted(os.listdir(train_images_dir))):
        img_path = os.path.join(train_images_dir, i)  # Correct concatenation
        img = cv2.imread(img_path)
        
        # Check if the image was loaded successfully
        if img is None:
            print(f"Warning: Unable to load image {img_path}. Skipping.")
            continue
        
        img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        img = 255 - img  # Creating the negative of the image
        img = cv2.resize(img, (128, 128))
        im.append(img)

    # Create DataFrame
    age = df_age['age']
    gen = df_gen['gender']
    data = {'Image': im, 'Age': age, 'Gender': gen}
    data_df = pd.DataFrame(data)
    print(data_df.head())


plt.imshow(cv2.cvtColor(data_df['Image'].iloc[0], cv2.COLOR_BGR2RGB))
plt.show()
X = np.array(data['Image'])
y = np.array(data['Gender'])

X = X.reshape((X.shape[0],X.shape[1],X.shape[2],1))
y = y.reshape(-1,1)
X = X/255
print(X.shape, y.shape)


import tensorflow
from tensorflow.keras.models import Sequential, Model, load_model
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Conv2D, MaxPool2D, Activation, BatchNormalization


def blockc(y,filt,kern):
    x = tensorflow.keras.layers.Conv2D(filt,kern, padding = 'same')(y)
    x = tensorflow.keras.layers.MaxPool2D(2 , strides = 2, padding = 'same')(x)
    x = tensorflow.keras.layers.BatchNormalization()(x)
    x = tensorflow.keras.layers.Activation('relu')(x)
    return x


#define model
tensorflow.keras.backend.clear_session()
model_gender = tensorflow.keras.Sequential()


inp = tensorflow.keras.layers.Input((128, 128,1))
c0 = tensorflow.keras.layers.Conv2D(4,3, padding = 'same')(inp)
c0 = tensorflow.keras.layers.MaxPool2D(2)(c0)
c1 = blockc(c0,16,3)
c2 = blockc(c1,32,3)
c3 = blockc(c2,64,3)
c4 = blockc(c3,128,3)
c5 = blockc(c4,64,3)
c6 = blockc(c5,32,3)
c6 =tensorflow.keras.layers.GlobalAveragePooling2D()(c6)
out = tensorflow.keras.layers.Dense(1,activation='sigmoid')(c6)
model_gender = tensorflow.keras.Model(inputs = [inp], outputs = [out])

model_gender.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model_gender.summary()


history = model_gender.fit(
    X,y,
    steps_per_epoch = (10000//10),
    validation_split=0.2,
    epochs = 50
)


pd.DataFrame(history.history).plot()

#model_gender.save('model_gender.h5')
model_gender.save('my_model.keras')  # Save using the .keras extension
y = np.array(data['Age'])
print(X.shape, y.shape)



tensorflow.keras.backend.clear_session()
model_age = tensorflow.keras.Sequential()


inp = tensorflow.keras.layers.Input((128, 128,1))
c0 = tensorflow.keras.layers.Conv2D(4,3, padding = 'same')(inp)
c0 = tensorflow.keras.layers.MaxPool2D(2)(c0)
c1 = blockc(c0,16,3)
c2 = blockc(c1,32,3)
c3 = blockc(c2,64,3)
c4 = blockc(c3,128,3)
c5 = blockc(c4,64,3)
c6 = blockc(c5,32,3)
c6 =tensorflow.keras.layers.GlobalAveragePooling2D()(c6)
c7 = tensorflow.keras.layers.Flatten()(c6)
c8 = tensorflow.keras.layers.Dense(100,activation='linear')(c7)
c9 = tensorflow.keras.layers.Dense(50,activation='linear')(c8)
out = tensorflow.keras.layers.Dense(1,activation='linear')(c9)
model_age = tensorflow.keras.Model(inputs = [inp], outputs = [out])


model_age.compile(loss='mean_squared_error', optimizer='adam', metrics=['mse'])
model_age.summary()

history = model_age.fit(
    X,y,
    steps_per_epoch = (10000//10),
    validation_split=0.2,
    epochs = 50
)

pd.DataFrame(history.history).plot()

model_age.save('my_model1.keras')

from sklearn.metrics import confusion_matrix,accuracy_score, precision_score, recall_score, f1_score

y_pred = model_gender.predict(X)
pred = []
for i in tqdm(y_pred):
    if i<0.5:
        pred.append(0)
    else:
        pred.append(1)
pred = np.array(pred)

print(pred.shape,y.shape)
print(pred[5:15],gen[5:15])


pred=np.round(model_age.predict(X))
print(pred[5:15],age[5:15])



import numpy as np
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
import seaborn as sns
import matplotlib.pyplot as plt

# Check the type and shape of the prediction array
print("Type of pred:", type(pred))
print("Shape of pred:", np.shape(pred))

# If pred is a 2D array, get the class with the highest probability
if len(pred.shape) > 1:  # If it's a 2D array (e.g., probabilities for each class)
    pred = np.argmax(pred, axis=1)  # Convert to class labels

# Check the distribution of classes in ground truth and predictions
print("Ground truth classes distribution:", np.bincount(gen))
print("Predicted classes distribution:", np.bincount(pred))

# Confusion matrix
cm = confusion_matrix(gen, pred)
print("Confusion Matrix:\n", cm)

# Plot the confusion matrix
sns.heatmap(cm, annot=True, fmt='g', cmap="Blues")
plt.ylabel('Prediction', fontsize=13)
plt.xlabel('Actual', fontsize=13)
plt.title('Confusion Matrix', fontsize=17)
plt.show()

# Accuracy
accuracy = accuracy_score(gen, pred)
print("Accuracy   :", accuracy)

# Precision (using 'weighted' average to handle class imbalance)
precision = precision_score(gen, pred, average='weighted')  # Change to 'weighted' for imbalanced classes
print("Precision :", precision)

# Recall (using 'weighted' average)
recall = recall_score(gen, pred, average='weighted')  # Change to 'weighted' for imbalanced classes
print("Recall    :", recall)

# F1-score (using 'weighted' average)
F1_score = f1_score(gen, pred, average='weighted')  # Change to 'weighted' for imbalanced classes
print("F1-score  :", F1_score)

# R2 score (assuming the model for 'age' is already defined and used)
from sklearn.metrics import r2_score
p = model_age.predict(X)
print("R2 Score:", r2_score(age, p))

from sklearn.metrics import r2_score

p = model_age.predict(X)

print(r2_score(age,p))

```
